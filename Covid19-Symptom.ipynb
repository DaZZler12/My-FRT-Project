```python
## Split the dataset for trainning and testing
```

```python
import numpy as np
```

```python
def dataset_split(data,ratio):
    np.random.seed(42)
    shuffled_val = np.random.permutation(len(data))
    testset_size = int(len(data) * ratio)
    test_indices = shuffled_val[:testset_size]
    train_indices = shuffled_val[testset_size:]
    
    return data.iloc[train_indices],data.iloc[test_indices]
```

```python
np.random.permutation(7)
data_frame
```

```python
train,test = dataset_split(data_frame,0.2)
```
